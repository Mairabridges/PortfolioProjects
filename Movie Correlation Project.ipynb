```python
# Import libraries

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)


# Read in the data

df = pd.read_csv(r'C:\Users\maira\Downloads\movies.csv')
```

```python
# The Data

df.head()
```

```python
# Missing Data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# Data Types for the columns

df.dtypes
```

```python
# Change data type of columns
df['budget'] = df['budget'].astype('Int64')
    
df['gross'] = df['gross'].astype('Int64')
```

```python
df.head()
```

```python
#Create correct year column
df['yearcorrect'] = df['year'].astype(str).str[:4]
```

```python
df.head()
```

```python
df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# Drop duplicates

df['company'].drop_duplicates().sort_values(ascending=False)
```

```python
df.head()
```

```python
# Scatter plot with Budget vs Gross


plt.scatter(x=df["budget"], y=df["gross"])

plt.title('Budget vs Gross Earnings')

plt.xlabel('Budget for Film')

plt.ylabel('Gross Earnings')

plt.show()
```

```python
# Plot budget vs gross using seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color": "red"}, line_kws={"color":"blue"})
```

```python
# Correlation

df.corr(method='pearson') 
```

```python
correlation_matrix = df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.scatter(x=df["budget"], y=df["gross"])

plt.title('Correlation Matric for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show
```

```python
# Company

df.head()
```

```python
df_numerical = df

for col_name in df_numerical.columns:
    if(df_numerical[col_name].dtype == 'object'):
        df_numerical[col_name] = df_numerical[col_name].astype('category')
        df_numerical[col_name] = df_numerical[col_name].cat.codes
        
df_numerical.head()
```

```python
correlation_matrix = df_numerical.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.scatter(x=df["budget"], y=df["gross"])

plt.title('Correlation Matric for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show

```

```python
df_numerical.corr()
```

```python
correlation_mat = df_numerical.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr
```
