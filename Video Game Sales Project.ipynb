```python
# Importing Libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

# Read in Data

df = pd.read_csv(r'C:\Users\maira\Downloads\vgsales.csv')
```

```python
df.head()
```

```python
# Data Types for Columns

df.dtypes
```

```python
# Change Data Type of Year, NA Sales, EU Sales, JP Sales, Other Sales, and Global Sales Columns

df['Year'] = df['Year'].astype('Int64')
```

```python
# Multiply Sales Columns to show Millions in Sales

df["NA_SalesMil"] = (df["NA_Sales"] * 1000000)

df["EU_SalesMil"] = (df["EU_Sales"] * 1000000)

df["JP_SalesMil"] = (df["JP_Sales"] * 1000000)

df["Other_SalesMil"] = (df["Other_Sales"] * 1000000)

df["Global_SalesMil"] = (df["Global_Sales"] * 1000000)
```

```python
df.head()
```

```python
# Delete old columns

df.drop('NA_Sales', axis=1, inplace=True)
```

```python

df.drop('EU_Sales', axis=1, inplace=True)

df.drop('JP_Sales', axis=1, inplace=True)

df.drop('Other_Sales', axis=1, inplace=True)

df.drop('Global_Sales', axis=1, inplace=True)
```

```python
df.head()
```

```python
# Sort by Year Ascending

df.sort_values(by=['Year'], inplace=False, ascending=True)
```

```python
# Top Selling Game in North America

df.sort_values(by=['NA_SalesMil'], inplace=False, ascending=False)
```

```python
# Top Selling Game in Japan

df.sort_values(by=['JP_SalesMil'], inplace=False, ascending=False)
```

```python
# Top Selling game Globally

df.sort_values(by=['Global_SalesMil'], inplace=False, ascending=False)
```

```python
# Top 5 Video Game Sales in North America

plt.scatter(x=df["Name"], y=df["NA_SalesMil"])

plt.xlim(0, 5)

plt.ylim(0,45000000)

plt.title('Top 5 North American Sales')

plt.xlabel('Name of Game')

plt.ylabel('North American Sales')

plt.show()
```

```python
# Top 5 Video Game Sales in Europe

plt.scatter(x=df["Name"], y=df["EU_SalesMil"])

plt.xlim(0, 5)

plt.ylim(0,45000000)

plt.title('Top 5 Europe Sales')

plt.xlabel('Name of Game')

plt.ylabel('Europe Sales')

plt.show()
```

```python
# Top 5 Video Game Sales in Japan

plt.scatter(x=df["Name"], y=df["JP_SalesMil"])

plt.xlim(0, 5)

plt.ylim(0,45000000)

plt.title('Top 5 Japan Sales')

plt.xlabel('Name of Game')

plt.ylabel('Japan Sales')

plt.show()
```

```python
# Top 5 Video Game Sales Globally

plt.scatter(x=df["Name"], y=df["Global_SalesMil"])

plt.xlim(0, 5)

plt.ylim(0,45000000)

plt.title('Top 5 of Global Sales')

plt.xlabel('Name of Game')

plt.ylabel('Global Sales')

plt.show()
```
